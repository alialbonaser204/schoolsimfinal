```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Laad de data
data = pd.read_csv("batch_results.csv")

# Splits per studentenaantal
students_30 = data[data["students"] == 30]
students_60 = data[data["students"] == 60]

# Functie om boxplots te maken met betrouwbaarheidsinterval


```

```python


def plot_metric(data, metric, title, ylabel):
    plt.figure(figsize=(8, 5))
    sns.boxplot(x="coffee_machines", y=metric, data=data, showfliers=False)

    # Gemiddelde + betrouwbaarheidsinterval (95%)
    means = data.groupby("coffee_machines")[metric].mean()
    cis = data.groupby("coffee_machines")[metric].agg(lambda x: 1.96 * x.std() / np.sqrt(len(x)))
    plt.errorbar(means.index, means, yerr=cis, fmt='o', color='red', label='Gemiddelde Â± CI')

    plt.title(title)
    plt.xlabel("Aantal koffieautomaten")
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

```

```python
# Wachttijd
plot_metric(students_30, "avg_wait_time", "Gemiddelde wachttijd - 30 studenten", "Wachttijd (minuten)")
plot_metric(students_60, "avg_wait_time", "Gemiddelde wachttijd - 60 studenten", "Wachttijd (minuten)")

```

```python
# Gemiste koffie
plot_metric(students_30, "missed_coffee", "Gemiste koffiepogingen - 30 studenten", "Aantal pogingen")
plot_metric(students_60, "missed_coffee", "Gemiste koffiepogingen - 60 studenten", "Aantal pogingen")

```

```python
# Idle-tijd
plot_metric(students_30, "avg_idle_time", "Gemiddelde idle-tijd - 30 studenten", "Idle tijd (minuten)")
plot_metric(students_60, "avg_idle_time", "Gemiddelde idle-tijd - 60 studenten", "Idle tijd (minuten)")
```

```python
def simple_boxplot(data, metric, title, ylabel):
    plt.figure(figsize=(8, 5))
    sns.boxplot(x="coffee_machines", y=metric, data=data, showfliers=False, palette="light:#5A9")
    plt.title(title)
    plt.xlabel("Aantal koffieautomaten")
    plt.ylabel(ylabel)
    plt.grid(True, axis='y')
    plt.tight_layout()
    plt.show()
```

```python
simple_boxplot(students_30, "avg_wait_time", "Wachttijd - 30 studenten", "Wachttijd (minuten)")
simple_boxplot(students_60, "avg_wait_time", "Wachttijd - 60 studenten", "Wachttijd (minuten)")

simple_boxplot(students_30, "missed_coffee", "Gemiste koffie - 30 studenten", "Aantal pogingen")
simple_boxplot(students_60, "missed_coffee", "Gemiste koffie - 60 studenten", "Aantal pogingen")

simple_boxplot(students_30, "avg_idle_time", "Idle tijd - 30 studenten", "Idle tijd (minuten)")
simple_boxplot(students_60, "avg_idle_time", "Idle tijd - 60 studenten", "Idle tijd (minuten)")
```

```python
def line_plot_with_ci(data, metric, title, ylabel):
    plt.figure(figsize=(8, 5))

    grouped = data.groupby("coffee_machines")[metric]
    means = grouped.mean()
    cis = grouped.agg(lambda x: 1.96 * x.std() / np.sqrt(len(x)))  # 95% CI

    plt.plot(means.index, means.values, marker='o', linestyle='-', color='blue', label='Gemiddelde')
    plt.fill_between(means.index, means - cis, means + cis, color='blue', alpha=0.2, label='95% CI')

    plt.title(title)
    plt.xlabel("Aantal koffieautomaten")
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
line_plot_with_ci(students_30, "avg_wait_time", "Gemiddelde wachttijd - 30 studenten", "Wachttijd (minuten)")
line_plot_with_ci(students_60, "avg_wait_time", "Gemiddelde wachttijd - 60 studenten", "Wachttijd (minuten)")
```
